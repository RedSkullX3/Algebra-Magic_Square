```python
import numpy as np
from prettytable import PrettyTable

# Function to display a matrix neatly
def display_matrix(matrix, title="Matrix"):
    table = PrettyTable()
    table.title = title
    for row in matrix:
        table.add_row(row)
    print(table)

# Function to compute A^k with square check
def matrix_power(matrix, k):
    rows, cols = matrix.shape
    if rows != cols:
        raise ValueError("Matrix must be square to compute powers.")
    return np.linalg.matrix_power(matrix, k)

# Function to check if a matrix is a magic square
def is_magic_square(matrix):
    n, m = matrix.shape
    if n != m:
        return False, None  # must be square
    
    target_sum = np.sum(matrix[0, :])  # sum of first row
    
    # Check rows
    for i in range(n):
        if np.sum(matrix[i, :]) != target_sum:
            return False, None
    
    # Check columns
    for j in range(n):
        if np.sum(matrix[:, j]) != target_sum:
            return False, None
    
    # Check diagonals
    if np.sum(np.diag(matrix)) != target_sum:
        return False, None
    if np.sum(np.diag(np.fliplr(matrix))) != target_sum:
        return False, None
    
    return True, target_sum

# Example usage
if __name__ == "__main__":
    # Define a sample n x n matrix
    A = np.array([
        [4, 3, 8],
        [9, 5, 1],
        [2, 7, 6]
    ])
    
    k = 3  # how many times to multiply
    
    display_matrix(A.tolist(), "Original Matrix A")
    
    try:
        Ak = matrix_power(A, k)
        display_matrix(Ak.tolist(), f"A^{k}")
        
        magic, magic_sum = is_magic_square(Ak)
        if magic:
            print(f"\n✅ A^{k} IS a magic square with magic sum = {magic_sum}")
        else:
            print(f"\n❌ A^{k} is NOT a magic square")
    except ValueError as e:
        print("Error:", e)
```

```python
def analyze_square(matrix):
    n = len(matrix)
    expected_sum = sum(matrix[0])

    print("Matrix:")
    for row in matrix:
        print(' '.join(f"{x:5}" for x in row))

    print("\nChecking conditions...")

    # Row sums
    row_sums = [sum(row) for row in matrix]
    rows_ok = all(r == expected_sum for r in row_sums)
    print(f"→ All row sums equal: {'✅' if rows_ok else '❌'} → {row_sums}")

    # Column sums
    col_sums = [sum(matrix[i][j] for i in range(n)) for j in range(n)]
    cols_ok = all(c == expected_sum for c in col_sums)
    print(f"→ All column sums equal: {'✅' if cols_ok else '❌'} → {col_sums}")

    # Diagonals
    main_diag = sum(matrix[i][i] for i in range(n))
    sec_diag = sum(matrix[i][n - 1 - i] for i in range(n))
    main_diag_ok = (main_diag == expected_sum)
    sec_diag_ok = (sec_diag == expected_sum)
    diagonals_equal = (main_diag == sec_diag)

    print(f"→ Main diagonal sum = {main_diag} → {'✅ matches' if main_diag_ok else '❌ does not match'}")
    print(f"→ Secondary diagonal sum = {sec_diag} → {'✅ matches' if sec_diag_ok else '❌ does not match'}")
    print(f"→ Diagonals are {'✅ equal' if diagonals_equal else '❌ not equal'}")

    print("\nFinal Result:")
    if rows_ok and cols_ok and main_diag_ok and sec_diag_ok:
        print("✅ This is a MAGIC SQUARE matrix.")
    elif rows_ok and cols_ok and diagonals_equal:
        print("⚠️ This is a SEMIMAGIC SQUARE matrix (diagonals are equal, but not fully magic).")
    elif rows_ok and cols_ok:
        print("ℹ️ This is a STRICTLY SEMIMAGIC SQUARE matrix (diagonals are not equal).")
    else:
        print("❌ This is neither a magic or semimagic square.")
#The following are indepent variales.
b=6
d=0
a=1
e=18
#the following variables are inconsequential or dependent variables
c=10
f=20
matrix = [
    [3*c*f-2*b*e-a*e-3*b*d, 3*c*f+b*e+2*a*e, 3*c*f+b*e-a*e+3*b*d],
    [3*c*f+b*e+2*a*e, 3*c*f-2*b*e-4*a*e, 3*c*f+b*e+2*a*e],
    [3*c*f+b*e-a*e+3*b*d, 3*c*f+b*e+2*a*e, 3*c*f-2*b*e-a*e-3*b*d]
]
analyze_square(matrix)

```

```python
#Making an SMS(3) matrix

import numpy as np
from prettytable import PrettyTable

# Function to display a matrix neatly
def display_matrix(matrix, title="Matrix"):
    table = PrettyTable()
    table.title = title
    for row in matrix:
        table.add_row(row)
    print(table)

# Function to build the matrix for given n and d
def build_matrix(n, d):
    matrix = np.array([
        [n - d,   n - 2*d, n + 3*d],
        [n + 4*d, n,       n - 4*d],
        [n - 3*d, n + 2*d, n + d]
    ])
    return matrix

# Example usage
if __name__ == "__main__":
    n = int(input("Enter value of n: "))
    d = int(input("Enter value of d: "))
    
    A = build_matrix(n, d)
    display_matrix(A.tolist(), f"Matrix for n={n}, d={d}")
```

```python
from itertools import combinations

def read_nine_ints():
    print("Enter 9 integers (space or newline separated):")
    data = []
    # Keep reading until we have 9 integers
    while len(data) < 9:
        try:
            line = input().strip()
        except EOFError:
            break
        if not line:
            continue
        parts = line.split()
        for p in parts:
            if len(data) >= 9:
                break
            try:
                data.append(int(p))
            except ValueError:
                print(f"Ignored non-integer token: {p}")
    return data

def try_magic(values):
    if len(values) != 9:
        print("Error: exactly 9 integers are required.")
        return

    total = sum(values)
    if total % 9 != 0:
        print("NO — sum of inputs is", total, "which is not divisible by 9. Cannot make magic square.")
        return

    c = total // 9
    if c not in values:
        print(f"NO — c = {c} (sum/9) does not appear among the input numbers. Cannot make magic square.")
        return

    sorted_vals = sorted(values)
    if sorted_vals[4] != c:
        print(f"NO — after sorting the 5th (median) value is {sorted_vals[4]}, not c = {c}. Cannot make magic square.")
        return

    
    for i, j in combinations(range(9), 2):
        vi = sorted_vals[i]
        vj = sorted_vals[j]
        a = vi - c
        b = vj - c

        L = [
            a + c,
            b + c,
            -a - b + c,
            -2*a - b + c,
            c,
            2*a + b + c,
            a + b + c,
            -b + c,
            -a + c
        ]

        if sorted(L) == sorted_vals:
            print("YES — a valid construction was found.")
            print(f"Used sorted positions {i+1} and {j+1} (values {vi} and {vj}) as the pair to form a and b.")
            print(f"Computed a = {a}, b = {b}, c = {c}.")
            print("\nThe 3×3 arrangement (row-wise, using the formula order):")
            print(f"[ {L[0]:>6}  {L[1]:>6}  {L[2]:>6} ]")
            print(f"[ {L[3]:>6}  {L[4]:>6}  {L[5]:>6} ]")
            print(f"[ {L[6]:>6}  {L[7]:>6}  {L[8]:>6} ]")
            return

    print("NO — checked all pairs; no valid construction found. Not possible to make the magic square with given numbers.")

if __name__ == "__main__":
    vals = read_nine_ints()
    try_magic(vals)
```
